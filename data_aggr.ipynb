```python
import numpy as np
import pandas as pd
```

```python
smoothness = pd.read_csv('1_Smoothness_Train.txt', sep='\s+', header=None)
leaflet_desc = pd.read_csv('7_Leaflet_Desc_Train.txt', sep='\s+', header=None)
gabor = pd.read_csv('8_Gabor_Train.txt', sep='\s+', header=None)
color = pd.read_csv('12_ColorTrain.txt', sep='\s+', header=None)
fft = pd.read_csv('50_FFT_Train.txt', sep='\s+', header=None)
geom = pd.read_csv('51_Geom_Train.txt', sep='\s+', header=None)
```

```python
features = ['smoothness', 'leaflet_desc', 'gabor', 'color', 'fft', 'geom']

for i in features:
    dataf_shape = globals()[i].shape[1] - 1
    new_cols = [i+str(x) for x in range(0,dataf_shape)]
    new_cols.append('code')
    globals()[i].columns = new_cols
```

```python
smoothness = smoothness[["code", "smoothness0"]]
```

```python
df_buffer = pd.concat([smoothness, leaflet_desc.iloc[:,:-1]], axis=1)

for i in features[2:]:
    df_buffer = pd.concat([df_buffer, globals()[i].iloc[:,:-1]], axis=1)
```
